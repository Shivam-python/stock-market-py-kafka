# Stock Market Analysis using kafka

```python
import random
import pandas as pd
import json
from time import sleep
from kafka import KafkaConsumer, KafkaProducer
```

```python
producer = KafkaProducer(
    bootstrap_servers='localhost:9092',
    value_serializer=lambda v: json.dumps(v).encode('utf-8')
)
```

```python
df = pd.read_csv("indexProcessed.csv")
```

```python
df.head()
```

```python
for i in range(100):
    data = df.sample(1).to_dict(orient="records")
    producer.send(
        topic='stock_data_v2',
        key=f"{random.randrange(999)}".encode(),
        value=data[0]
    )
    print(f"Sent: {data[0]}",end="\n\n")
```

```python
producer.flush()
```
